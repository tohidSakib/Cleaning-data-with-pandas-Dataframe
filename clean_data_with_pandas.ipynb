## importing libraries

```python
import pandas as pd
import numpy as np
```

## getting data

```python
data = pd.read_csv('Table of Size Standards_Effective July 14 2022.csv')
```

```python
data.head()
```

```python
data.info()
```

## checking for nulls

```python
data.isnull().sum()
```

```python
data = data[data['NAICS Codes'].isnull()]
```

```python
data.sample(10)
```

```python
data.info()
```

```python
data['Lookup Codes'] = data['NAICS Industry Description'].str.slice(start=7, stop=9)
```

```python
data['Sector'] = data['NAICS Industry Description'].str.slice(start=12, stop=538)
```

```python
data
```

```python
data[['NAICS Industry Description', 'Lookup Codes', 'Sector']]
```

```python
data['Sector'] = np.where(data['Sector'] == '33 – Manufacturing', 'Manufacturing', data['Sector'])
```

```python
data['Sector'] = np.where(data['Sector'] == '45 – Retail Trade', 'Retail Trade', data['Sector'])
```

```python
data['Sector'] = np.where(data['Sector'] == '49 – Transportation and Warehousing', 'Transportation and Warehousing', data['Sector'])
```

```python
data.head(10)
```

```python
data = data[['NAICS Industry Description', 'Lookup Codes', 'Sector']]
```

```python
data = data[data['Lookup Codes'].str.isnumeric()]
```

```python
data.head()
```

```python
data.loc[len(data.index)] = ['Sector 31 - 33 - Manufacturing', 32, 'Manufacturing']
data.loc[len(data.index)] = ['Sector 31 - 33 - Manufacturing', 33, 'Manufacturing']
```

```python
df = {'NAICS Industry Description': 'Sector 44 - 45 – Retail Trade', 'Lookup Codes': 45, 'Sector': 'Retail Trade'}
data = data.append(df, ignore_index =  True)
```

```python
data.sample(10)
```

```python
data.loc[len(data.index)] = ['Sector 48 - 49 – Transportation and Warehousing', 49, 'Transportation and Warehousing']
```

```python
data.head()
```

### Now time to create csv file with clean data
